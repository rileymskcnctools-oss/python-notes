```python
def printme(str):
    print(str)
    return
    
   
# 调用函数
printme("我要调用用户自定义函数!")
printme("再次调用同一函数")

def ChangeInt( a ):
    a = 10
 
b = 2
ChangeInt(b)
print(b) # 整数是不可变对象


def ChangeList(list):
    list.append('web')
    print("改变前的列表：",list)
    return
    
 
list1=[1,2,3,'app']
ChangeList(list1)
print("改变后的列表：",list1) # 列表是可变对象

def Changetup(tup):
    # 这里 tup 是 tup1 的一个引用
    print("函数内部（操作前）：", tup)
    
    # 我们不能修改元组的内容，也不能通过 del tup 删除外部变量
    # 如果在这里 del tup，只是删除了函数内的这个变量名
    tup = ('new', 'tuple') # 这只是让局部变量 tup 指向了一个新对象
    print("函数内部（赋值后）：", tup)
    return

tup1 = ('physics', 'chemistry', 1997, 2000)

print("调用前：", tup1)
Changetup(tup1)
print("调用后：", tup1)


#可写函数说明
def printinfo( name, age ):
   "打印任何传入的字符串"
   print ("Name: ", name)
   print ("Age: ", age)
   return
 
#调用printinfo函数
printinfo("miki",50 )  #
    
```

```python
# 定义一个变量
PI = 3.14159

# 定义函数 1：打招呼
def say_hello(name):
    """这是一个打印问候语的函数"""
    print(f"你好, {name}！欢迎学习 Python 模块。")

# 定义函数 2：计算圆面积
def circle_area(radius):
    """输入半径，返回圆的面积"""
    area = PI * (radius ** 2)
    return area

# 注意：通常模块里只放定义，不放直接运行的代码
```

```python
import random  # 随机数模块
import time    # 时间模块

print("正在生成随机数...")
time.sleep(1)  # 程序暂停1秒

# 生成 1 到 100 之间的随机整数
lucky_num = random.randint(1, 100)
print(f"你的幸运数字是: {lucky_num}")

# 从列表中随机选一个
fruits = ['苹果', '香蕉', '橘子']
choice = random.choice(fruits)
print(f"今晚吃: {choice}")
```

```python
import math

# 1. 使用我们在 dir() 中看到的常量
print("圆周率 pi:", math.pi)
print("自然常数 e:", math.e)

# 2. 使用我们在 dir() 中看到的函数
# 使用 sqrt (square root) 计算平方根
print("16 的平方根:", math.sqrt(16))

# 使用 ceil (ceiling) 向上取整
print("3.14 向上取整:", math.ceil(3.14))

# 使用 pow (power) 计算乘方
print("2 的 3 次方:", math.pow(2, 3))
```
