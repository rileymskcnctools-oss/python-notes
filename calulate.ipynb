```python
%%writefile my_calc.py

# 文件名: my_calc.py (注意：这一行会被 %%writefile 写入文件)

def add(x, y):
    """相加"""
    return x + y

def subtract(x, y):
    """相减"""
    return x - y

def multiply(x, y):
    """相乘"""
    return x * y

def divide(x, y):
    """相除 (带有除以0的检查)"""
    if y == 0:
        return "错误：除数不能为0"
    return x / y

# 运行 Cell 后，你会在输出看到 "Writing my_calc.py"
```

```python
# 导入我们刚刚创建的模块
import my_calc

# 1. 调用加法函数
result_add = my_calc.add(50, 25)
print(f"50 + 25 = {result_add}")

# 2. 调用除法函数 (测试正常情况)
result_div = my_calc.divide(100, 20)
print(f"100 / 20 = {result_div}")

# 3. 测试除以零的错误处理
error_msg = my_calc.divide(10, 0)
print(f"10 / 0 = {error_msg}")
import importlib

# 强制 Python 重新加载 my_calc 模块
importlib.reload(my_calc)

# 再次调用函数，这次它会使用最新的代码
print("重新加载后，再次计算 50 + 25:")
print(f"50 + 25 = {my_calc.add(50, 25)}")

```
