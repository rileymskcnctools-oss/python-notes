```python
import numpy as np

arr = np.array([1, 2, 3, 4])

print(type(arr))      # 查看类型
print(arr.shape)      # 数组形状
print(arr.dtype)      # 数据类型
```

```python
# 从列表创建
arr1 = np.array([1, 2, 3])

# 创建全 0 数组
arr2 = np.zeros(9)

# 创建全 1 数组
arr3 = np.ones(8)

# 创建连续数字
arr4 = np.arange(0, 12, 2)
print(arr1)
print(arr2)
print(arr3)
print(arr4)
```

```python
arr = np.array([1, 2, 3, 4])

print(arr + 10)
print(arr * 2)
print(arr.mean())
print(arr.sum())
print(arr.max())
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

print(arr.shape)  #数组的属性
print(arr.ndim)      # 维度
print(arr.size)     # 元素个数
print(arr.dtype)    # 数据类型
print(arr.mean())  #数组方法--求平均值
print(arr.std())  #数组方法--标准差
print(arr * 5)   #数组运算
print(arr[-3:-1])   #数组切片
print(np.sqrt(arr))     # 开平方
print(np.log(arr))      # 取对数
print(np.exp(arr))      # 指数

```
