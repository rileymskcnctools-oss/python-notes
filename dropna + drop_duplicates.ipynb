```python
import pandas as pd
import numpy as np

data = {
    "订单号": ["A001", "A002", "A003", "A003", "A004", None],
    "客户": ["张三", "李四", None, None, "王五", "赵六"],
    "金额": [100, None, 300, 300, 500, 200],
    "国家": ["中国", "美国", "中国", "中国", None, "日本"]
}

df = pd.DataFrame(data) #使用字典创建数据帧，键为列，默认索引为整数
print(f"创建新的数据帧是：\n{df}")
#df_clean1=df.dropna(subset=["金额"])  #删除特定列为空的行
#print(f"删除金额为空的行————————————————————-新的数据帧是：\n{df_clean1}")

#df_clean2=df.dropna(subset=["订单号"])  #删除特定列为空的行
#print(f"删除订单号为空的行————————————————————-新的数据帧是：\n{df_clean2}")
#df_clean3=df.drop_duplicates(subset=["订单号"]) #按列去重，保留第一个
#print(f"删除重复订单————————————————————新的数据帧是：\n{df_clean3}")
#df_clean4=df.drop_duplicates(subset=["订单号"],keep="last") #保留最后一个重复项
#print(f"删除重复订单————————————————————新的数据帧是：\n{df_clean4}")

print("---------------df_fill---未知客户填充以后-------------------")
df['客户']=df['客户'].fillna("未知客户")
print(df)
print("---------------df_fill---金额空值填充以后-------------------")
df['金额']=df['金额'].fillna(df['金额'].mean())
print(df)
print("---------------df_fill---国家空值填充以后-------------------")
df['国家'] = df['国家'].fillna(method="ffill")
print(df)
print("---------------df_fill---剩余空值填充以后-------------------")
df_fill=df.fillna("缺失")
print(df_fill)
```
