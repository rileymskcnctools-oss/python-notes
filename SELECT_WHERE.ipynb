```python
import pandas as pd
from pandasql import sqldf

# 1. 加载 CSV 文件到 DataFrame
df = pd.read_csv('onlineretail.csv')

# 2. 定义一个快捷函数，方便后续运行 SQL
# globals() 使得 sqldf 能够识别当前 Notebook 中的 DataFrame 变量名 (如 df)
pysqldf = lambda q: sqldf(q, globals())
q1 = """
SELECT  InvoiceNo, StockCode, Quantity, Country
FROM df
WHERE Quantity=6;
"""
result1 = pysqldf(q1)
print(result1)
q2 = """
SELECT InvoiceNo, StockCode, Quantity, Country
FROM df
WHERE Quantity > 10
  AND Country = 'United Kingdom';
"""
result2 = pysqldf(q2)
print("\n--- 练习一结果 (Quantity > 10 AND Country='United Kingdom') ---")
print(result2.head())
```

```python
import pandas as pd
from pandasql import sqldf

# 1. 加载 CSV 文件到 DataFrame
df = pd.read_csv('onlineretail.csv')

# 2. 定义一个快捷函数，方便后续运行 SQL
# globals() 使得 sqldf 能够识别当前 Notebook 中的 DataFrame 变量名 (如 df)
pysqldf = lambda q: sqldf(q, globals())
q1 = """
SELECT  *
FROM df;
"""
result1 = pysqldf(q1)
print(f"编写 SQL 语句，选出表中所有产品的所有列信息{result1}")
q2 = """
SELECT Description,UnitPrice
FROM df;
"""
result2 = pysqldf(q2)
print(f"编写 SQL 语句，选出表中Description,UnitPrice列信息\n{result2}")
q3 = """
SELECT DISTINCT Country 
FROM df;
"""
result3 = pysqldf(q3)
print(f"编写 SQL 语句，选出表中不重复的国家-列信息\n{result3}")
q4= """
SELECT * 
FROM df
WHERE Quantity>50;
"""
result4 = pysqldf(q4)
print(f"编写 SQL 语句，选出所有购买数量 Quantity 大于 50 的交易记录的所有信息\n{result4}")
q5= """
SELECT * 
FROM df
WHERE Country='France' AND UnitPrice<5 ;
"""
result5 = pysqldf(q5)
print(f"编写 SQL 语句，选出所有 Country 为 'France' 并且 UnitPrice 小于 5 的交易记录\n{result5}")

q6= """
SELECT * 
FROM df
WHERE Country!='United Kingdom' AND Quantity=1 ;
"""
result6 = pysqldf(q6)
print(f"编写 SQL 语句，选出所有 Country 不是 'United Kingdom' 并且 Quantity 等于 1 的交易记录\n{result6}")
q7= """
SELECT * 
FROM df
WHERE Country IN ('Germany','Norway' , 'Sweden') ;
"""
result7 = pysqldf(q7)
print(f"编写 SQL 语句，选出所有 Country 是 'Germany'、'Norway' 或 'Sweden' 的交易记录。请使用 IN 操作符\n{result7}")
```
