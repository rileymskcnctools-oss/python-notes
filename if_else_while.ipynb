```python
for num in range(20,30):  # 迭代 20 到 30 (不包含) 之间的数字
   for i in range(2,num): # 根据因子迭代
      if num%i == 0:      # 确定第一个因子
         j=num/i          # 计算第二个因子
         print ('%d 等于 %d * %d' % (num,i,j))
         break            # 跳出当前循环，已确定不是质数，所以继续下一轮外层循环
   else:                  # for–else 的 else 只有在 for 循环“没有被 break”时才执行
      print ('%d 是一个质数' % num)
```

```python
i = 2
while i < 100:
    j = 2
    while j*j< i:
        if i % j == 0:
            break
        j += 1
    else:
        print(i, "是素数")
    i += 1

print("Good bye!")
```

```python
# ===== 1. 创建字典 =====
student = {
    "name": "Tom",
    "age": 18,
    "class": "Grade 3"
}

print("原始字典：", student)


# ===== 2. 查（读取）=====
print("\n--- 查 ---")
print("姓名：", student["name"])
print("年龄：", student.get("age"))
print("性别（不存在返回None）：", student.get("gender"))


# ===== 3. 增（新增）=====
print("\n--- 增 ---")
student["gender"] = "Male"
student["score"] = 95
print("添加后的字典：", student)


# ===== 4. 改（修改）=====
print("\n--- 改 ---")
student["age"] = 20
student["score"] = 98
print("修改后的字典：", student)


# ===== 5. 删（删除）=====
print("\n--- 删 ---")
del student["class"]
print("删除 class 后：", student)

score = student.pop("score")
print("删除的 score：", score)
print("当前字典：", student)


# ===== 6. 遍历 =====
print("\n--- 遍历 ---")
for key in student:
    print(key, "=>", student[key])

print("---------")

for k, v in student.items():
    print(k, "=>", v)


# ===== 7. 判断键是否存在 =====
print("\n--- 判断键是否存在 ---")
if "name" in student:
    print("name 键存在")

if "class" not in student:
    print("class 键不存在")


# ===== 8. 获取字典长度 =====
print("\n--- 字典长度 ---")
print("字典长度：", len(student))
```

```python
# ===== 1. 创建字典 =====
customer={
    "name":"Green Tools Ltd",
    'country':"Brazil",
    'order_num':1500,
    'price':9.9   
}

# ===== 2. 查（读取）=====
print("\n--- 查 ---")
print(f"客户名：{customer['name']},国家:{customer['country']},单价:{customer['price']}")

print("\n--- 增 ---")
customer["contact"] = "WhatsApp"
customer["payment"] = "TT"
print("添加后的字典：", customer)
# ===== 4. 改（修改）=====
print("\n--- 改 ---")
customer['country']='Paraguay'
customer['price']=8.8
print("修改后的字典：", customer)
# ===== 5. 删（删除）=====
print("\n--- 删 ---")
del customer["payment"]
print("删除 payment 后：", customer)
# ===== 6. 计算订单总金额=====
order_val=customer['price']*customer['order_num']
print("订单总金额：", order_val)
# ===== 7. 遍历====
print("\n---遍历 ---")
print("客户信息如下：")
for key in customer:
    print(key, "=>", customer[key])
print("---------")
# ===== 8. 判断字段是否存在====
if 'contact' in customer:
    print("有联系方式")
if 'contact' not in customer:
    print("没有联系方式")
for k, v in customer.items():
    print("key:", k, type(k))
    print("value:", v, type(v))
    print("------")

```
